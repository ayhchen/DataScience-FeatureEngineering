```python
max_threshold = df['price'].quantile(0.95)
max_threshold
```

```python
min_threshold = df['price'].quantile(0.05)
min_threshold
```

```python
df_fixed = df[(df.price > min_threshold) & (df.price < max_threshold)]
df_fixed.shape
```
